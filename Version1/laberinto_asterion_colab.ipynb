```python
# ============================================================
# CLASE PRINCIPAL DEL JUEGO
# ============================================================

class LaberintoAsterion:
    """Clase principal del juego El Laberinto de Asteri√≥n"""
    
    def __init__(self, configuracion):
        """Inicializa el juego con la configuraci√≥n proporcionada"""
        self.config = configuracion
        self.tablero = None
        self.posicion = None
        self.jugador = None
        self.personajes_en_tablero = {}
        self.inicializar_juego()
    
    def inicializar_juego(self):
        """Inicializa el tablero, jugador y personajes"""
        # Crear tablero 10x10
        tamano = 10
        self.tablero = [[None for _ in range(tamano)] for _ in range(tamano)]
        
        # Posici√≥n inicial del jugador (centro del laberinto)
        self.posicion = [tamano // 2, tamano // 2]
        
        # Inicializar estad√≠sticas del jugador desde configuraci√≥n
        stats = self.config['estadisticas_base_jugador']
        self.jugador = {
            'nombre': stats['nombre'],
            'salud': stats['salud'],
            'cordura': stats['cordura'],
            'sigilo': stats['sigilo'],
            'velocidad': stats['velocidad'],
            'puntos': stats['puntos'],
            'inventario': [],
            'hilo_ariadna': stats['hilo_ariadna_inicial']
        }
        
        # Colocar personajes aleatoriamente en el tablero
        self.colocar_personajes()
    
    def colocar_personajes(self):
        """Coloca los 5 personajes aliados en posiciones aleatorias del tablero"""
        personajes = self.config['personajes_aliados']
        coordenadas = self.generar_posiciones_aleatorias(10, len(personajes))
        
        for personaje, coord in zip(personajes, coordenadas):
            x, y = coord
            # Guardamos el ID del personaje en el tablero
            self.tablero[x][y] = personaje['id']
            # Guardamos la informaci√≥n completa del personaje
            self.personajes_en_tablero[personaje['id']] = personaje
    
    def generar_posiciones_aleatorias(self, n, k):
        """Genera k posiciones √∫nicas aleatorias en el tablero nxn"""
        coords = set()
        while len(coords) < k:
            coords.add((random.randrange(n), random.randrange(n)))
        return list(coords)
    
    def mostrar_titulo(self):
        """Muestra el t√≠tulo del juego"""
        info = self.config['info_juego']
        lore = self.config['contexto_lore']
        
        print("\n" + "="*60)
        print(f"  {info['titulo'].upper()}")
        print(f"  {info['subtitulo']}")
        print("="*60)
        print(f"\n[*] Ubicaci√≥n: {lore['ubicacion']}")
        print(f"[!] Protagonista: {lore['protagonista']}")
        print(f"[X] Enemigo: {lore['villano_principal']}")
        print(f"\n[>] Objetivo: {self.config['objetivo_principal']}")
        print("\n" + "="*60 + "\n")
    
    def mostrar_estado_jugador(self):
        """Muestra las estad√≠sticas actuales del jugador"""
        print(f"\n{'‚îÄ'*60}")
        print(f"[@] {self.jugador['nombre']} | ", end="")
        print(f"[HP] Salud: {self.jugador['salud']} | ", end="")
        print(f"[CD] Cordura: {self.jugador['cordura']} | ", end="")
        print(f"[--] Hilo: {self.jugador['hilo_ariadna']} | ", end="")
        print(f"[**] Puntos: {self.jugador['puntos']}")
        inventario_texto = ', '.join(self.jugador['inventario']) if self.jugador['inventario'] else 'Vac√≠o'
        print(f"[##] Inventario ({len(self.jugador['inventario'])}/5): {inventario_texto}")
        print(f"{'‚îÄ'*60}\n")
    
    def verificar_condiciones_derrota(self):
        """Verifica si el jugador ha perdido"""
        if self.jugador['salud'] <= 0:
            print("\n[XX] Tu salud ha llegado a 0. Has muerto en el laberinto...")
            return True
        
        if self.jugador['cordura'] <= 0:
            print("\n[@@] Tu cordura se ha desvanecido. Te pierdes en la locura del laberinto...")
            return True
        
        if self.jugador['hilo_ariadna'] <= 0:
            print("\n[--] El Hilo de Ariadna se ha agotado. Est√°s perdido para siempre...")
            return True
        
        return False
    
    def encontrar_personaje(self, personaje_id):
        """Maneja el encuentro con un personaje aliado"""
        personaje = self.personajes_en_tablero[personaje_id]
        
        print(f"\n{'='*60}")
        print(f"[+] ¬°Has encontrado a {personaje['nombre']}!")
        print(f"{'='*60}")
        print(f"\n[i] {personaje['descripcion']}")
        print(f"\n[~] {personaje['nombre']}: \"{personaje['dialogo']}\"\n")
        
        # Mostrar opciones de objetos
        objetos = personaje['objetos']
        print("[?] Objetos disponibles:\n")
        
        for idx, obj in enumerate(objetos, 1):
            print(f"  {idx}) {obj['nombre']}")
            print(f"     Tipo: {obj['tipo']}")
            print(f"     {obj['descripcion']}")
            
            # Mostrar estad√≠sticas espec√≠ficas del objeto
            if 'dano' in obj:
                print(f"     [ATK] Da√±o: {obj['dano']}")
            if 'defensa' in obj:
                print(f"     [DEF] Defensa: {obj['defensa']}")
            if 'habilidad' in obj:
                print(f"     [SKL] Habilidad: {obj['habilidad']}")
            if 'poder' in obj:
                print(f"     [PWR] Poder: {obj['poder']}")
            print()
        
        # Solicitar elecci√≥n del jugador
        while True:
            try:
                eleccion = int(input("Elige un objeto (1-3): "))
                if 1 <= eleccion <= 3:
                    objeto_elegido = objetos[eleccion - 1]
                    break
                else:
                    print("[!] Opci√≥n inv√°lida. Elige 1, 2 o 3.")
            except ValueError:
                print("[!] Por favor ingresa un n√∫mero.")
        
        # Agregar objeto al inventario
        self.jugador['inventario'].append(objeto_elegido['nombre'])
        print(f"\n[OK] Has obtenido: {objeto_elegido['nombre']}")
        
        # Remover personaje del tablero
        x, y = self.posicion
        self.tablero[x][y] = None
        
        # Aplicar efecto aleatorio
        self.aplicar_efecto_aleatorio()
    
    def aplicar_efecto_aleatorio(self):
        """Aplica un efecto aleatorio despu√©s de recoger un objeto"""
        efectos = self.config['efectos_aleatorios']
        efecto = random.choice(efectos)
        
        self.jugador[efecto['tipo']] += efecto['valor']
        
        print(f"\n[?] Efecto aleatorio: {efecto['descripcion']}")
        
        # Mostrar cambio
        signo = '+' if efecto['valor'] > 0 else ''
        print(f"   {efecto['tipo'].capitalize()}: {signo}{efecto['valor']}")
    
    def mover_jugador(self):
        """Solicita y procesa el movimiento del jugador"""
        print("\n[^] Direcciones disponibles:")
        print("   N (Norte) | S (Sur) | E (Este) | O (Oeste)")
        
        movimiento = input("\nElige direcci√≥n: ").upper().strip()
        
        x, y = self.posicion
        tamano = 10
        movimiento_valido = False
        
        if movimiento == 'N' and x > 0:
            self.posicion[0] -= 1
            movimiento_valido = True
        elif movimiento == 'S' and x < tamano - 1:
            self.posicion[0] += 1
            movimiento_valido = True
        elif movimiento == 'E' and y < tamano - 1:
            self.posicion[1] += 1
            movimiento_valido = True
        elif movimiento == 'O' and y > 0:
            self.posicion[1] -= 1
            movimiento_valido = True
        else:
            print("[!] Movimiento inv√°lido. No puedes salir del laberinto.")
            return
        
        if movimiento_valido:
            # Consumir hilo de Ariadna
            self.jugador['hilo_ariadna'] -= 1
            print(f"[OK] Te has movido hacia el {movimiento}")
            print(f"   [--] Hilo de Ariadna restante: {self.jugador['hilo_ariadna']}")
    
    def determinar_final(self):
        """Determina qu√© final obtiene el jugador seg√∫n su inventario"""
        inventario_conjunto = set(self.jugador['inventario'])
        finales = self.config['finales_victoria']
        
        print("\n" + "="*60)
        print("  [!!!] ENFRENTAMIENTO FINAL CON ASTERI√ìN")
        print("="*60 + "\n")
        
        print(f"Con tu inventario completo, te adentras al centro del laberinto...")
        print(f"Objetos recolectados: {', '.join(self.jugador['inventario'])}\n")
        
        input("Presiona ENTER para enfrentar al Minotauro...")
        
        # Buscar final v√°lido
        final_encontrado = None
        for final in finales:
            combinacion_requerida = set(final['combinacion'])
            if inventario_conjunto == combinacion_requerida:
                final_encontrado = final
                break
        
        print("\n" + "="*60)
        
        if final_encontrado:
            print(f"  [***] {final_encontrado['nombre']}")
            print("="*60)
            print(f"\n{final_encontrado['descripcion']}\n")
            print(f"[>] {final_encontrado['narrativa']}")
            print(f"\n[WIN] ¬°VICTORIA! Has derrotado a Asteri√≥n y escapado del laberinto.")
        else:
            # Final de derrota
            final_derrota = self.config['final_derrota']
            print(f"  [XXX] {final_derrota['nombre']}")
            print("="*60)
            print(f"\n{final_derrota['descripcion']}\n")
            print(f"[>] {final_derrota['narrativa']}")
            print(f"\n[XXX] DERROTA. El laberinto reclama otra v√≠ctima...")
        
        print("\n" + "="*60)
        print(f"  Puntuaci√≥n Final: {self.jugador['puntos']} puntos")
        print("="*60 + "\n")
    
    def jugar(self):
        """Bucle principal del juego"""
        self.mostrar_titulo()
        
        print("Comienzas tu aventura en el centro del laberinto...")
        print("Debes encontrar a los 5 aliados dispersos y recolectar sus objetos.\n")
        
        input("Presiona ENTER para comenzar...")
        
        # Bucle principal: hasta recolectar 5 objetos o morir
        while len(self.jugador['inventario']) < 5:
            # Verificar condiciones de derrota
            if self.verificar_condiciones_derrota():
                print("\n[XXX] GAME OVER")
                return
            
            # Mostrar estado
            self.mostrar_estado_jugador()
            
            x, y = self.posicion
            print(f"[*] Posici√≥n actual: Sala ({x}, {y})")
            
            # Verificar si hay personaje en esta celda
            personaje_id = self.tablero[x][y]
            if personaje_id:
                self.encontrar_personaje(personaje_id)
            else:
                print("   Esta sala est√° vac√≠a. Contin√∫a explorando...")
            
            # Mover jugador
            self.mover_jugador()
        
        # Final del juego
        self.determinar_final()

print("‚úÖ Clase del juego cargada correctamente")
```
